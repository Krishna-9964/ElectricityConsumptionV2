```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("data.csv")
dataset
```

```python
df = pd.DataFrame(dataset)
df
```

```python
dataset.info()
```

```python
df[['Day','Month','Year']] = df.Dates.str.split("-", expand = True)
df
```

```python
df['Year'] = df['Year'].str.replace(' 00:00', '')
df
```

```python
df.drop(['Dates'], axis=1)
```

```python
def get_state_usage():
    # filtered_df = df[(df['States'] == state) & (df['Year'] == year)]
    # filtered_df.sort_values("Month",ascending=False)
    usage_sum =  df.groupby('States')['Usage'].sum().reset_index()
    sorted_usage = usage_sum.sort_values('Usage', ascending=False).head(5)
    json_data = sorted_usage.to_json(orient='records')
    return json_data

k = get_state_usage()
k


```

```python
def predict_consumption(State,Month,Day):
    state_index = np.where(x_train.columns==State)[0][0]
    k = np.zeros(len(x_train.columns))
    k[0] = Month
    k[1] = Day
    if state_index >=0 :
        k[state_index] = 1
        return reg.predict([k])[0]
```
